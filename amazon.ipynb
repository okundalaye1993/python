## Which Year maximum number of forest fires were reported 


```python
# Group the data by year and calculate the total number of fires
yearly_fires = df.groupby('year')['number'].sum()

# Find the year with the maximum number of fires
max_year = yearly_fires.idxmax()

max_year
```

```python
### visualize
```

```python
# Visualize the data
plt.figure(figsize=(10, 6))
yearly_fires.plot(kind='bar', color='skyblue')
plt.title('Total Number of Forest Fires by Year')
plt.xlabel('Year')
plt.ylabel('Total Number of Fires')
plt.xticks(rotation=45)
plt.annotate(f'Max: {yearly_fires[max_year]:.0f}', (max_year, yearly_fires[max_year]), xytext=(6, 50),
             textcoords='offset points', arrowprops=dict(arrowstyle="->"))

plt.show()
```

### Average number of fires reported from highest to lowest(state-wise) and visualize

```python
# Calculate the average number of fires reported for each state
state_avg_fires = df.groupby('state')['number'].mean().reset_index()

# Sort the states by the average number of fires from highest to lowest
state_avg_fires = state_avg_fires.sort_values(by='number', ascending=False)

# Visualize the data with a bar chart
plt.figure(figsize=(10, 6))
plt.bar(state_avg_fires['state'], state_avg_fires['number'], color='skyblue')
plt.title('Average Number of Fires Reported by State (Highest to Lowest)')
plt.xlabel('State')
plt.ylabel('Average Number of Fires')
plt.xticks(rotation=45)
```
